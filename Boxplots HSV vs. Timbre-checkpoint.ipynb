```python
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import colorsys

# load the data from a CSV file into a DataFrame
M2I = pd.read_csv('Data - I2M.csv')
I2M = pd.read_csv('Data - M2I.csv')

timbre_order = ['Bowed Strings', 'Percussion', 'Woodwind', 'Brass', 'Plucked Strings']

# calculate the mean HSV values for each timbre class M2I
M2I_grouped = M2I.groupby('Timbre Class')
M2I_avg = M2I_grouped.mean()[['Hue', 'Saturation', 'Value']]

# convert the mean HSV values to RGB values for each timbre class M2I
M2I_colors = {}
for timbre in timbre_order:
    hsv = M2I_avg.loc[timbre].values / [360, 1, 1]
    rgb = colorsys.hsv_to_rgb(*hsv)
    M2I_colors[timbre] = rgb

# calculate the mean HSV values for each timbre class I2M
I2M_grouped = I2M.groupby('Timbre Class')
I2M_avg = I2M_grouped.mean()[['Hue', 'Saturation', 'Value']]

# convert the mean HSV values to RGB values for each timbre class I2M
I2M_colors = {}
for timbre in timbre_order:
    hsv = I2M_avg.loc[timbre].values / [360, 1, 1]
    rgb = colorsys.hsv_to_rgb(*hsv)
    I2M_colors[timbre] = rgb

# create a box plot of the Hue variable for each timbre class M2I
sns.boxplot(x='Timbre Class', y='Hue', data=M2I, order=timbre_order, palette=M2I_colors)
plt.xlabel('Timbre Class')
plt.ylabel('Hue')
plt.title('M2I Hue for All Timbre Classes')
plt.show()

# create a box plot of the Saturation variable for each timbre class M2I
sns.boxplot(x='Timbre Class', y='Saturation', data= M2I, order=timbre_order, palette= M2I_colors)
plt.xlabel('Timbre Class')
plt.ylabel('Saturation')
plt.title('M2I Saturation for All Timbre Classes')
plt.show()

# create a box plot of the Value variable for each timbre class M2I
sns.boxplot(x='Timbre Class', y='Value', data= M2I, order=timbre_order, palette=M2I_colors)
plt.xlabel('Timbre Class')
plt.ylabel('Value')
plt.title('M2I Value for All Timbre Classes')
plt.show()

# create a box plot of the Hue variable for each timbre class I2M
sns.boxplot(x='Timbre Class', y='Hue', data=I2M, order=timbre_order, palette=I2M_colors)
plt.xlabel('Timbre Class')
plt.ylabel('Hue')
plt.title('I2M Hue for All Timbre Classes')
plt.show()

# create a box plot of the Saturation variable for each timbre class I2M
sns.boxplot(x = 'Timbre Class', y= 'Saturation', data= I2M, order = timbre_order, palette=I2M_colors)
plt.xlabel('Timbre Class')
plt.ylabel('Saturation')
plt.title('I2M Saturation for All Timbre Classes')
plt.show()

# create a box plot of the Value variable for each timbre class I2M
sns.boxplot(x = 'Timbre Class', y= 'Value', data= I2M, order = timbre_order, palette=I2M_colors)
plt.xlabel('Timbre Class')
plt.ylabel('Value')
plt.title('I2M Value for All Timbre Classes')
plt.show()







```
