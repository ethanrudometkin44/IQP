```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# read the CSV file into a DataFrame
df = pd.read_csv('Data - I2M.csv')

# select the columns of interest
cols = ['RR', 'DV', 'TempoNum', 'Hue', 'Saturation', 'Value']

# create a list of timbre classes
timbre_classes = df['Timbre Class'].unique()

# create subplots for each timbre class
fig, axes = plt.subplots(nrows=len(timbre_classes), figsize=(6, 12))

# loop over each timbre class and plot the correlation matrix
for i, timbre_class in enumerate(timbre_classes):
    # filter the data on the current timbre class
    filtered_df = df[df['Timbre Class'] == timbre_class]
    data = filtered_df[cols]
    corr_matrix = data.corr()
    # create a heatmap of the correlation matrix
    sns.heatmap(corr_matrix, ax=axes[i], cmap='coolwarm', vmin=-1, vmax=1, annot=True, fmt='.2f')
    # set the title of the subplot
    axes[i].set_title(timbre_class)

# adjust the spacing between subplots
fig.tight_layout()

# display the plot
plt.show()
```
