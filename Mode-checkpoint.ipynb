```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
import colorsys

# read the CSV file into a DataFrame
df = pd.read_csv('Data - M2I.csv')
filtered_df = df[df['Mode'] == 'Minor']

# create the figure and 3D axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# create RGB color values for each point
colors = []
for i in range(len(filtered_df)):
    hsv = (filtered_df['Hue'].iloc[i]/360, filtered_df['Saturation'].iloc[i], filtered_df['Value'].iloc[i])
    rgb = colorsys.hsv_to_rgb(*hsv)
    colors.append(rgb)

# scatter plot with Hue, Saturation, and Brightness as x, y, and z axes
ax.scatter(filtered_df['Hue'], filtered_df['Saturation'], filtered_df['Value'], c=colors, alpha=0.5)

# set labels for each axis
ax.set_xlabel('Hue')
ax.set_ylabel('Saturation')
ax.set_zlabel('Brightness')

# set title for the plot
ax.set_title('M2I 3D Scatter Plot of Hue, Saturation, and Brightness (Minor)')

# get number of points
num_points = len(filtered_df)
print("Number of points:", num_points)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
import colorsys

# read the CSV file into a DataFrame
df = pd.read_csv('Data - M2I.csv')
filtered_df = df[df['Mode'] == 'Minor']

# create a dictionary to count color wheel colors
color_wheel_count = {
    'red': 0,
    'orange': 0,
    'yellow': 0,
    'yellow-green': 0,
    'green': 0,
    'cyan-green': 0,
    'cyan': 0,
    'blue-cyan': 0,
    'blue': 0,
    'blue-violet': 0,
    'violet': 0,
    'red-violet': 0,
}

# iterate over each point in the filtered_df DataFrame
for i in range(len(filtered_df)):
    hue = filtered_df['Hue'].iloc[i]
    if 0 <= hue <= 30:
        color_wheel_count['red'] += 1
    elif 30 < hue <= 60:
        color_wheel_count['orange'] += 1
    elif 60 < hue <= 90:
        color_wheel_count['yellow'] += 1
    elif 90 < hue <= 120:
        color_wheel_count['yellow-green'] += 1
    elif 120 < hue <= 150:
        color_wheel_count['green'] += 1
    elif 150 < hue <= 180:
        color_wheel_count['cyan-green'] += 1
    elif 180 < hue <= 210:
        color_wheel_count['cyan'] += 1
    elif 210 < hue <= 240:
        color_wheel_count['blue-cyan'] += 1
    elif 240 < hue <= 270:
        color_wheel_count['blue'] += 1
    elif 270 < hue <= 300:
        color_wheel_count['blue-violet'] += 1
    elif 300 < hue <= 330:
        color_wheel_count['violet'] += 1
    elif 330 < hue <= 360:
        color_wheel_count['red-violet'] += 1

# print the counts of color wheel colors
for color, count in color_wheel_count.items():
    print(f"{color.capitalize()}: {count}")
    
# create a dictionary to map color names to colors for the bar graph
color_wheel_colors = {
    'red': 'red',
    'orange': 'orange',
    'yellow': 'yellow',
    'yellow-green': 'chartreuse',
    'green': 'green',
    'cyan-green': 'springgreen',
    'cyan': 'cyan',
    'blue-cyan': 'deepskyblue',
    'blue': 'blue',
    'blue-violet': 'violet',
    'violet': 'magenta',
    'red-violet': 'crimson',
}

# create a list of color names and counts in the order we want them to appear on the bar graph
color_names = list(color_wheel_count.keys())
color_counts = [color_wheel_count[color_name] for color_name in color_names]

# create a dictionary to map color names to their abbreviated form
color_abbreviations = {
    'red': 'R',
    'orange': 'O',
    'yellow': 'Y',
    'yellow-green': 'YG',
    'green': 'G',
    'cyan-green': 'CG',
    'cyan': 'C',
    'blue-cyan': 'BC',
    'blue': 'B',
    'blue-violet': 'BV',
    'violet': 'V',
    'red-violet': 'RV',
}

# create a list of abbreviated color names for display on the bar graph
color_names_abbreviated = [color_abbreviations[color_name] for color_name in color_names]

# create a bar graph with colored bars and abbreviated color names
fig, ax = plt.subplots()
ax.bar(color_names_abbreviated, color_counts, color=[color_wheel_colors[color_name] for color_name in color_names])

# set labels and title for the plot
ax.set_xlabel('Color')
ax.set_ylabel('Count')
ax.set_title('M2I Bar Graph of Color Count (Minor)')

# show the plot
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# read the CSV file into a DataFrame
df = pd.read_csv('Data - M2I.csv')
filtered_df = df[df['Mode'] == 'Major']

# create RGB color values for each point
colors = []
for i in range(len(filtered_df)):
    hsv = (filtered_df['Hue'].iloc[i]/360, filtered_df['Saturation'].iloc[i], filtered_df['Value'].iloc[i])
    rgb = colorsys.hsv_to_rgb(*hsv)
    colors.append(rgb)

# create a scatter plot with Saturation and Brightness as x and y axes
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(filtered_df['Saturation'], filtered_df['Value'], c=colors, alpha=0.5)
ax.set_xlabel('Saturation')
ax.set_ylabel('Brightness')
ax.set_title('Title')

# add a median line for the Saturation values
median_saturation = filtered_df['Saturation'].median()
ax.axvline(x=median_saturation, linestyle='--', color='gray')

# add a median line for the Brightness values
median_brightness = filtered_df['Value'].median()
ax.axhline(y=median_brightness, linestyle='--', color='gray')

# show the plot
plt.show()

# get number of points
num_points = len(filtered_df)
print("Number of points:", num_points)
```
