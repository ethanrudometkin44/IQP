```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as signal
import numpy as np
import pandas as pd

# Load the existing CSV file
df = pd.read_csv('M2I.csv')

# Loop through audio files
for i in range(1, 16):
    # Load audio file
    audio_file = f'{i}.wav'
    y, sr = librosa.load(audio_file)

    # calculate power spectral density (PSD)
    f, psd = signal.welch(y, fs=sr, nperseg=2048)

    # calculate spectral centroid
    spectral_centroid = np.sum(f * psd) / np.sum(psd)

    # Find the row where Trial is i and store the spectral centroid
    df.loc[df['Trial'] == i, 'Spectral Centroid'] = spectral_centroid

# Save the modified DataFrame back to the CSV file
df.to_csv('M2I.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the existing CSV file
df = pd.read_csv('M2I.csv')

# Create scatter plot
plt.scatter(df['Spectral Centroid'], df['Value'])
plt.xlabel('Spectral Centroid')
plt.ylabel('Value')
plt.show()
```
